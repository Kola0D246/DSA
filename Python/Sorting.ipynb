```python
#This function need to be edit for multiple iterations and for sublist

def checkPivot(lst, pivot):
  '''This function checks if all the leftside element of pivot are smaller and all the rightside element of pivot are larger than pivot'''

  #check element on right
  inPivot = lst.index(pivot)
  left = inPivot
  right = len(lst)-1
  for x in range(right,left,-1):
    if lst[x] > lst[inPivot]:
      return False

  #check element on left
  inPivot = lst.index(pivot)
  left = 0
  right = inPivot
  for x in range(left, right):  #step = 1 is default value
    if lst[x] < lst[inPivot]:
      return False

  return True


def Qsort(lst, pivot):
  '''This function is for quick sort
  lst = target list
  pivot = value that we are using as pivot'''

  #check if pivot value exist
  if pivot not in lst: return "Pivot value not found"

  #check for smaller element from right
  inPivot = lst.index(pivot)      #store index of pivot in inPivot
  left = inPivot                  #we are using left and right variable to make code easily readable.
  right = len(lst)-1              #we can simply put the inPivot and len(lst)-1 as argument in range() as well
  for x in range(right,left,-1):
    if lst[x] < lst[inPivot]:
      lst[x], lst[inPivot] = lst[inPivot], lst[x]

  #check for larger element from left
  inPivot = lst.index(pivot)
  left = 0
  right = inPivot
  for x in range(left, right):  #step = 1 is default value
    if lst[x] > lst[inPivot]:
      lst[x], lst[inPivot] = lst[inPivot], lst[x]

  #if checkPivot(lst, pivot) == False: Qsort(lst, pivot) #Recursion happens until pivot is at correct position
  return lst
```

```python
# Quick sort in Python

# function to find the partition position
def partition(array, low, high):

  # choose the rightmost element as pivot
  pivot = array[high]

  # pointer for greater element
  i = low - 1

  # traverse through all elements
  # compare each element with pivot
  for j in range(low, high):
    if array[j] <= pivot:
      # if element smaller than pivot is found
      # swap it with the greater element pointed by i
      i = i + 1

      # swapping element at i with element at j
      (array[i], array[j]) = (array[j], array[i])

  # swap the pivot element with the greater element specified by i
  (array[i + 1], array[high]) = (array[high], array[i + 1])

  # return the position from where partition is done
  return i + 1

# function to perform quicksort
def quickSort(array, low, high):
  if low < high:

    # find pivot element such that
    # element smaller than pivot are on the left
    # element greater than pivot are on the right
    pi = partition(array, low, high)

    # recursive call on the left of pivot
    quickSort(array, low, pi - 1)

    # recursive call on the right of pivot
    quickSort(array, pi + 1, high)


data = [8, 7, 2, 1, 0, 9, 6]
print("Unsorted Array")
print(data)

size = len(data)

quickSort(data, 0, size - 1)

print('Sorted Array in Ascending Order:')
print(data)
```

##Insertion Sort

<font size=5> Algorithm </font>

1. Consider actual list into two sublist: Sorted and Unsorted sublist
2. Pick the first element of unsorted sublist and compare it with each element of sorted sublist until a bigger element is found.
  - Direction of Iteration: Left -> Right
3. Move the picked element - remove from unsorted sublist and insert in sorted sublist on left of the bigger element.
4. Repeat step 2 and 3 for all the elements in unsorted list
5. Termination after unsorted list ends. The actual list is sorted.

Initially, the whole list is unsorted and the sorted sublist is empty. So we take the first element of actual list as sorted sublist because there is nothing to compare.

##Bubble Sort

- Use in Array with less number of element; small dataset
- Time Consuming
- Simple Code

<font size = 5> Algorithm <font>

1. Compare first two elements of table and swap them if the elements are in wrong order
2. Repeat step 1 for next pair (the new right element and element next to it), until the end of array is reached. This ends first pass
3. Repeat step 1 and 2 until list is sort.

```python
##My code
def BubbleSort(arr):
  sort = False
  while not sort:
    sort = True         #Chat GPT Note: Variable name should be small unless they are constant
    print("Pass")
    for x in range(len(arr)-1):
      if arr[x] > arr[x+1]:
        arr[x], arr[x+1] = arr[x+1], arr[x]
        print(arr)              #for debugging
        sort = False
  return arr

arr = [7,6,100,4,3,5]
BubbleSort(arr)

print(arr)

#Time Complexity = O()
#Space Complexity = O()
```

##Selection Sort

<font size=5> Algorithm <font>

1. Start from leftmost element and compare it to next element in linear order until an element smaller than it is found
2. Take new element and comapre it with next elements until an element smaller than it is found. Repeat this, selecting the smallest element by the end of list
3. Swap the smallest element found with the first element of unsorted portion
4. Repeat above steps for remaining portion of unsorted sublist until whole list is sorted

##Merge Sort

```python
# Function to merge two halves
def merge(left, right):
    result = []
    i = j = 0

    # Compare elements of both halves and merge them into a sorted result
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Append remaining elements of left and right
    result.extend(left[i:])
    result.extend(right[j:])

    return result

# Function to implement Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Split the list into two halves
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    # Recursively sort each half and merge them
    left = merge_sort(left)
    right = merge_sort(right)

    # Merge the sorted halves
    return merge(left, right)

# Example usage

arr = [38, 27, 43, 3, 9, 82, 10]
print("Original array:", arr)
sorted_array = merge_sort(arr)
print("Sorted array:", sorted_array)
```
