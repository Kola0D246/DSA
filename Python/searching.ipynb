##Binary Search

Searching algorithm that
1. Sort the list
2. Divide search space in two halves by the middle index called mid
3. Compare the mid value with desired (key) value:
  - if key is at mid, then process is terminated and index of mid value is return
  - if key is not at mid, then choose which side of mid is used as next search space:
    - if key is smaller than mid value, then select left side of mid
    - if key is larger than mid value, then select right side of mid
4. This process continues until the key is found or no search space is left (the previous search space is single element ie mid value)

```python
#BINARY SEARCH

def Bsearch(lst, key):          #This code takes less time and memory campare to actually slicing the list
    #sort the list
    lst.sort()


    low = 0
    high = len(lst) -1

    while True:
        mid = (low + high)//2
        if key == lst[mid]: return mid  #return index of desired element
        elif key < lst[mid]: high = mid -1
        elif key > lst[mid]: low = mid+1
    return None                 #If desired element is not in list

lst_raw = [4, 6, 0, 2, 10, 9, 1]
key = 6
print("Index of desired element:", Bsearch(lst_raw, key))
```

##Linear Search

Searching Algorithm that compares key value with every element of list in sequence of indexing until:
1. key value - terminate the process and return the current index.
2. list ends without finding the key element - return -1 as the desired element is not in the list

```python
def Lsearch(lst, key):
  for x in range(len(lst)):
    if key == lst[x]: return x
  return -1
```

```python
lst = [4, 6, 0, 2, 10, 9]
key = 6
print("Index of desired element:", Lsearch(lst, key))
```
