```python
class node:
  def __init__(self, value, double= False):
    self.value= value
    self.next= None
    if double:
      self.prev= None

class singlyLinkedList:
  def __init__(self):
    self.head= None
    self.size= 0

  def findNode(self, position):
    self.ER_outofbound(position, self.size)
    cursor = self.head
    for i in range(1, position):
      cursor = cursor.next
    return cursor
    
  def insert(self, value, position=0):
    self.ER_outofbound(position, self.size)
    new = node(value)
    if position==0:
      cursor = self.head
      self.head = new
    else:
      prev = self.findNode(position-1)
      cursor = prev.next
      prev.next = new
    new.next = cursor
    self.size+=1
  
  def delete(self, position):
    self.ER_outofbound(position, self.size-1)
    if position==0:
      cursor=self.head
      self.head=cursor.next
    else:
      prev = self.findNode(position-1)
      cursor = prev.next
      prev.next = cursor.next
    del cursor
    self.size-=1

  def display(self, frm, to):
    self.ER_outofbound(frm, self.size)
    self.ER_outofbound(to, self.size)
    if frm>to: raise ValueError("Invalid range")

    cursor = self.findNode(frm)
    for i in range(frm, to):
      print(f'[{cursor.value}] -> ', end='')
      cursor = cursor.next
    print('END')

  def ER_outofbound(self, position, size):
    if position <0 or position > size:
      raise ValueError("Linked list out of bound")
```
