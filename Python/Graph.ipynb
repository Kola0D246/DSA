# GRAPH

<font size=5> Theory </font>

Graph are non-linear data structure consist of vertices and edges.
1. Vertices: aka Nodes, are points in graph that store some value
2. Edges are line connecting two nodes
  - Directed Edge have direction: Node1 -> Node2. <br>
     Can tranverse only in specified direction
  - Undirected Edge have no direction: Node1 - Node2. <br>
     Can tranverse in both direction
  - Weighted: Each edge has value {cost, weight}
  - Unweighted: Simple edge

> Graph Representation in dictionary

1. object = graph representation stored as dictionary <br>
2. keys = individual vertecies of graph <br>
3. value of key = list of vertices connected to key by an edge <br>

For Directed Edge: Update value of only first argumented key <br>
For Undirected Edge: Update value of both argumented keys <br>

```python
def dfs(graph, node, visited):
    if node not in visited:
        print(node, end=" ")  # Process the node
        visited.add(node)  # Mark node as visited
        for neighbor in graph.get(node, []):
            dfs(graph, neighbor, visited)
  # Explore neighbors

# Example graph as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

visited = set()
print("DFS Traversal:")
dfs(graph, 'A', visited)  # Start DFS from node 'A'
```

```python
def bfs(graph, start):
    visited = set()
    queue = [start]  # Using a list as a queue (FIFO)

    while queue:
        node = queue.pop(0)  # Dequeue the first element
        if node not in visited:
            print(node, end=" ")  # Process the node
            visited.add(node)  # Mark node as visited
            queue.extend(graph.get(node, []))
# Enqueue all unvisited neighbors

# Example graph as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("BFS Traversal:")
bfs(graph, 'A')  # Start BFS from node 'A'
```
